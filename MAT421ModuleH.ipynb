**22.1 ODE Iniitial Value Problem Statement**
*   A differential Equation is a relstionship between a function, its independent variable and any number of its derivatives
*   An Ordinary Differential Equation (ODE) is one where the independent variable and the derivatives are in one dimension
*   Partial Differential Equations are a general form of a differential equation where the independent variable is a vector containing $(x_1, x_2, ..., x_m)$ and the partial derivatives can be any order and with respect to any combination of variables
*   A general solution is a function that satisfies the differential equation
*   A particular solution is a function that satisfies the differential equation and n known values of the solution or derivatives at certain points
*   An initial value problem involves finding the solution to an ODE given an initial value.



**22.2 Reduction of Order**
*   Many methods of solving IVPs are meant for first order differential equations. When working with higher order differential equations, reducing their order is an important first step
*   The state of a system is a collection of all relevant dependent variables
*   Taking the derivative of the state of a system reduces the order of a differential equation to the first order
*   A linear ODE is on that can be written in matrix form



**22.3 The Euler Method**
*   The explicit Euler Formula allows for the approximation for the state at $S(t_j+1)$ given the state at $S(t_j)$ using $S(t_j+1)=S(t_j)+h\cdot F(t_j,S(t_j))$
*   Given an initial value for $S_0$, the formula can be used to integrate the steps up to $S(t_f)$
*   Using a smaller step size (h) will provide more accurate approximations

```python
import numpy as np
import matplotlib.pyplot as plt

f = lambda t, s: np.exp(-t)
h = 0.1
t = np.arange(0, 1 + h, h)
s0 = -1

s = np.zeros(len(t))
s[0] = s0

for i in range(0, len(t) - 1):
    s[i + 1] = s[i] + h*f(t[i], s[i])

plt.figure(figsize = (12, 8))
plt.plot(t, s, 'bo--', label='Approximate') ; plt.plot(t, -np.exp(-t), 'g', label='Exact')
plt.title('Approximate and Exact Solution for Simple ODE with Step Size of 0.1')
plt.xlabel('t') ; plt.ylabel('f(t)')
plt.grid()      ; plt.legend(loc='lower right')
plt.show()
```
