```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d
from scipy.interpolate import CubicSpline
```

**17.1 Interpolation Problem Satement**
*   Interpolation is a process used to estimate a value between two points which have predetermined values.
*   Interpolation can be simpler than normal regression as it creates a model between each set of points, rather than a more complex model to fit all the points.




```python
x = [1, 2, 3, 4, 5, 6, 7]
y = [3, 5, 4, 7, 4.5, 8, 7]

plt.figure(figsize = (3,3))
plt.plot(x, y, 'o')
plt.ylim(0, 10)
plt.title('Graph of Data')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

**17.2 Linear Interpolation**
*   Linear interpolation creates a line between consecutive points and assumes that the a point in between will lie on that line.

```python
# Using the same data from earlier
x = [1, 2, 3, 4, 5, 6, 7]
y = [3, 5, 4, 7, 4.5, 8, 7]

# Using interpolation to estimate the point at x = 5.5
f = interp1d(x,y)
y_hat = f(5.5)

plt.figure(figsize = (5,5))
plt.plot(x, y, '-o')
plt.plot(5.5, y_hat, 'ro')
plt.ylim(0, 10)
plt.title('Linear Interpolation at x = 5.5')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

print(f(5.5))
```

**17.3 Cubic Spline Interpolation**
*   Cublic spline interpolation creates a set of piecewise cubic functions to connect each data point rather than a set of lines.
*   For a set of n data points, n-1 cubic functions must be created, and each function requires 4 coefficients.

```python
# Using the same data from earlier
x = [1, 2, 3, 4, 5, 6, 7]
y = [3, 5, 4, 7, 4.5, 8, 7]

# Using interpolation to estimate the point at x = 5.5
f = CubicSpline(x, y, bc_type='natural')
x_new = np.linspace(1, 7, 100)
y_new = f(x_new)

plt.figure(figsize = (5,5))
plt.plot(x_new, y_new, 'b')
plt.plot(x, y, 'ro')
plt.ylim(0, 10)
plt.title('Cubic Spline Interpolation')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

print(f(5.5))
```
